# Treasure Hunt

```python
def disp():
    print("Treasure Hunt\n")
    print("You are on a deserted island that is used by pirates\n\to hide their treasure. The pirates are coming back\n\
    soon, and you only have a short amount of time to find\n\the treasure before they return and catch you trying\n\
    to steal from them.")

    print("")

    print("Here's a map of the island - each '.' represents a spot\n\where you can dig for treasure. When prompted, enter\n\
    the ROW and COLUMN you wish to dig. You have 3 chances\n\to dig before the pirates return!")
    print("")

    print(" " * 11 + " NORTH")
    print(" " * 8 + "1 2 3 4 5 6 7")
    print(" " * 6 + "1 " + ". " * 7)
    print(" " * 6 + "2 " + ". " * 7)
    print(" " * 6 + "3 " + ". " * 7)
    print("WEST  " + "4 " + ". " * 7 + " EAST")
    print(" " * 6 + "5 " + ". " * 7)
    print(" " * 6 + "6 " + ". " * 7)
    print(" " * 6 + "7 " + ". " * 7)
    print(" " * 11 + " SOUTH")


def hint(r, c, row, col):
    if row < r:
        print("Go SOUTH")

    if row > r:
        print("Go NORTH")

    if col < c:
        print("Go EAST")
    if col > c:
        print("Go WEST")


disp()

import random

r = random.choice([1, 2, 3, 4, 5, 6, 7])
c = random.choice([1, 2, 3, 4, 5, 6, 7])

flag = True
if flag:
    print("")
    print("Guess #1")
    row = int(input("Enter a ROW: "))
    col = int(input("Enter a COLUMN: "))
    print("")
    if row == r and col == c:
        flag = False
        print("You found the treasure on Guess #1!  Congratulations!")
    else:
        print("No treasure here! Here are some hints for your next guess:")
        hint(r, c, row, col)


if flag:
    print("")
    print("Guess #2")
    row = int(input("Enter a ROW: "))
    col = int(input("Enter a COLUMN: "))
    print(" ")
    if row == r and col == c:
        flag = False
        print("You found the treasure on Guess #2!  Nice job!")
    else:
        print("No treasure here! Here are some hints for your next guess:")
        hint(r, c, row, col)

if flag:
    print("")
    print("Guess #3")
    row = int(input("Enter a ROW: "))
    col = int(input("Enter a COLUMN: "))
    print(" ")
    if row == r and col == c:
        flag = False
        print("You found the treasure on Guess #3!  Just in time!")
    else:
        print("No treasure here, and the pirates have returned to the island! You've been caught!")
        print("The treasure was at ROW " + str(r) + " and COLUMN " + str(c))
```

# Date Checker

```python
def leap(year):
    if year % 100 == 0:
        if year % 400 == 0:
            return True

        else:
            return False

    else:
        if year % 4 == 0:
            return True

        else:
            return False


def letter(day):
    if day == 1 or day == 21 or day == 31:
        return "st "
    elif day == 2 and day == 22:
        return "nd "
    elif day == 3 and day == 23:
        return "rd "
    else:
        return "th "


date = input("Enter a date ( YYYYMMDD ): ")
year = int(date[:4])
month = int(date[4:6])
day = int(date[6:])

mname = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November",
         "December"]
months = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
flag = True

if month in range(1, 13):
    flag = False
if leap(year):
    print(str(year) + " is a leap year")
else:
    print(str(year) + " is NOT a leap year")

if month == 2:
    if (leap(year) and day < 30) or not (leap(year)) and day < 29:
        print("February " + str(day) + letter(day) + str(year) + " is a valid date ")
    else:
        print("This is not a valid date in " + str(year))
else:
    if month < 13 and month != 0 and months[month - 1] >= day:
        print(mname[month - 1] + " " + str(day) + letter(day) + str(year) + " is a valid date")
    else:
        print("This is not a valid date in " + str(year))
```
