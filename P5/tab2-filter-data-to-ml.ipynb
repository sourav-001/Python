<div class="alert alert-block alert-info">
<b>NOTE: confusion matrix :</b><br>
    <i>confusion matrix is a tabular summary of the number of correct and incorrect predictions made by a classifier.<br>
       It is used to measure the performance of a classification model. It can be used to evaluate the performance of a classification model through the calculation of performance metrics like accuracy, precision, recall, and F1-score. <br>
       For an imbalanced dataset to work with, it’s always better to use confusion matrix as my evaluation criteria for my machine learning model.<br>
       We can measure model accuracy by two methods. Accuracy simply means the number of values correctly predicted.<br>
       1. Confusion Matrix<br>
       2. Classification Measure<br>
       <u>1. Confusion Matrix:</u>
        - True Positives (TP): when the actual value is Positive and predicted is also Positive.<br>
- True negatives (TN): when the actual value is Negative and prediction is also Negative.<br>
- False positives (FP): When the actual is negative but prediction is Positive. Also known as the Type 1 error<br>
- False negatives (FN): When the actual is Positive but the prediction is Negative. Also known as the Type 2 error<br>
        
  ![Confusion matrix](https://miro.medium.com/max/1334/1*3yGLac6F4mTENnj5dBNvNQ.jpeg)
 <br>
        <u>2. Classification Measure</u><br>
Basically, it is an extended version of the confusion matrix. There are measures other than the confusion matrix which can help achieve better understanding and analysis of our model and its performance.<br>
a. Accuracy<br>
b. Precision<br>
c. Recall (TPR, Sensitivity)<br>
d. F1-Score<br>
e. FPR (Type I Error)<br>
f. FNR (Type II Error)<br>
        <br>
        
        
        a. Accuracy:
Accuracy simply measures how often the classifier makes the correct prediction. It’s the ratio between the number of correct predictions and the total number of predictions. The accuracy metric is not suited for unbalanced classes.<br>
Accuracy has its own disadvantages, for imbalanced data, when the model predicts that each point belongs to the majority class label, the accuracy will be high. But, the model is not accurate.
        <br>
        ![Accuracy](https://miro.medium.com/max/700/1*ICP55HLZ6xxXaUysUU8s2w.jpeg)
        <br>
        
        b. Precision:
Precision is defined as the ratio of the total number of correctly classified positive classes divided by the total number of predicted positive classes. Or, out of all the predictive positive classes, how much we predicted correctly. Precision should be high(ideally 1).<br>
“Precision is a useful metric in cases where False Positive is a higher concern than False Negatives”<br>
        
 ![Precision](https://miro.medium.com/max/585/1*vzCv0_sP9S0OE07wA9wIEQ.jpeg)<br>
        
        
        
        c. Recall:
Recall is defined as the ratio of the total number of correctly classified positive classes divide by the total number of positive classes. Or, out of all the positive classes, how much we have predicted correctly. Recall should be high(ideally 1).<br>
“Recall is a useful metric in cases where False Negative trumps False Positive”
        <br>
        ![Recall](https://miro.medium.com/max/680/1*QU9h_Ye6YGrs6wtE94V3WQ.jpeg)
        <br>
        4. F-measure / F1-Score
There will be cases where there is no clear distinction between whether Precision is more important or Recall. We combine them!<br>
In practice, when we try to increase the precision of our model, the recall goes down and vice-versa. The F1-score captures both the trends in a single value.<br>
        
 ![F1-Score](https://miro.medium.com/max/611/1*WLCBBeQHeFoKdOwZP7Wm-w.jpeg)
        <br>
       
        F1 score is a harmonic mean of Precision and Recall. As compared to Arithmetic Mean, Harmonic Mean punishes the extreme values more. F-score should be high(ideally 1)<br>
        5. Sensitivity & Specificity
![Sensitivity & Specificity](https://miro.medium.com/max/657/1*FJNxbk7Ccr1hINGQmgKMcw.jpeg)
        <br>
        <b>When to use Accuracy / Precision / Recall / F1-Score?<b><br>
            a. Accuracy is used when the True Positives and True Negatives are more important. Accuracy is a better metric for Balanced Data.<br>
b. Whenever False Positive is much more important use Precision.<br>
c. Whenever False Negative is much more important use Recall.<br>
d. F1-Score is used when the False Negatives and False Positives are important. F1-Score is a better metric for Imbalanced Data.<r>
            
</div>

## class weight balance for DTC
___________________________

```python
# class weight balance
for_decisiontree = DecisionTreeClassifier(class_weight = "balanced")
model_dtc = for_decisiontree.fit(X_train, y_train)
```

```python
# function for computing confusion matrix and scores

def evalute_for_model_dtc(which_model, X_test_data, y_test_data):
    print("DTC")
    model_accuracy = which_model.score(X_test_data, y_test_data)
    print("Test Accuracy: {:.2f}%".format(model_accuracy * 100))
    
    y_ture = np.array(y_test_data)
    y_pred = which_model.predict(X_test_data)
    
    confusionMatriX = confusion_matrix(y_ture, y_pred)
    #print(f"\n----------------------\nConfusion Matrix :\n----------------------\n{confusionMatriX}")
    classificationReporT = classification_report(y_ture, y_pred, target_names=['BENIGN', 'MALIGNANT'])
    
    # for graph ploting
    plt.figure(figsize=(10, 10))
    sns.heatmap(confusionMatriX, annot=True, vmin=0, fmt='g', cmap='Blues', cbar=False)
    plt.xticks(np.arange(2) + 0.5, ["BENIGN", "MALIGNANT"])
    plt.yticks(np.arange(2) + 0.5, ["BENIGN", "MALIGNANT"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()
    print(f"\n----------------------\nClassification Report:\n----------------------\n{classificationReporT}")
```

```python
# running function
evalute_for_model_dtc(for_decisiontree, X_test, y_test)
```

<div class="alert alert-block alert-info">
<b>Standardization:</b><br>
    caling of Features is an essential step in modeling the algorithms with the datasets. The data that is usually used for the purpose of modeling is derived through various means such as:<br>
    Questionnaire, Surveys, Research, Scraping, etc.<br>
    So, the data obtained contains features of various dimensions and scales altogether. Different scales of the data features affect the modeling of a dataset adversely.

It leads to a biased outcome of predictions in terms of misclassification error and accuracy rates. Thus, it is necessary to Scale the data prior to modeling.
    <br>
    <b>Standardization is a scaling technique wherein it makes the data scale-free by converting the statistical distribution of the data into the below format:<b>
        <br>
        -mean - 0<br>
        -standard dviation - 1<br>
        ![SD](https://cdn.journaldev.com/wp-content/uploads/2020/10/Standardization.png.webp)
        
        
        <br>
  
        By this, the entire data set scales with a zero mean and unit variance, altogether.<br>
 
        Have to import StandardScaler from sklearn.preprocesing<br>
 __from sklearn.preprocessing import StandardScaler__
   create and object of StanderdScaler.
sc = StandardScaler()
__Note: Standardization is only applicable on the data values that follows Normal Distribution__
        
        

</div>

___________________

__The idea behind StandardScaler is that it will transform your data such that its distribution will have a mean value 0 and standard deviation of 1.
In case of multivariate data, this is done feature-wise (in other words independently for each column of the data).
Given the distribution of the data, each value in the dataset will have the mean value subtracted, and then divided by the standard deviation of the whole dataset (or feature in the multivariate case).__


<b>fit_transform()</b><br>
__fit_transform() is used on the training data so that we can scale the training data and also learn the scaling parameters of that data. Here, the model built by us will learn the mean and variance of the features of the training set. These learned parameters are then used to scale our test data.__<br>
_The fit method is calculating the mean and variance of each of the features present in our data. The transform method is transforming all the features using the respective mean and variance._
